```python
import json
with open('output.json','r') as f:
    output =  json.load(f)
    
print(output)
```

```python
import pandas as pd
import numpy as np
%matplotlib inline
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
import plotly.express as px
import plotly.graph_objects as go
```

```python
loss = output['loss']
steps = output['step']
wer = output['cer']
cer = output['wer']
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=steps, y=wer,
                    mode='lines+markers',
                    name='WER'))
fig.add_trace(go.Scatter(x=steps, y=cer,
                    mode='lines+markers',
                    name='CER'))

fig.update_layout(
    title="Error Rates vs Iterations",
    xaxis_title="STEPS",
    yaxis_title="Error Rate",
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="#7f7f7f"
    )
)
fig.show()
#fig.write_image("ErrorRatevsIterations.png")
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=steps, y=loss,
                    mode='lines+markers',
                    name='LOSS'))

fig.update_layout(
    title="Val Loss vs Iterations",
    xaxis_title="STEPS",
    yaxis_title="Loss",
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="#7f7f7f"
    )
)
fig.show()
#fig.write_image("ErrorRatevsIterations.png")
```
